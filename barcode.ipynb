```python
import barcode
from barcode.writer import ImageWriter

#specify the barcode format
barcode_type=barcode.get_barcode_class("ean13")

#create barcode object with 12-digit number
barcode_number="987654321012"
barcode_obj=barcode_type(barcode_number, writer=ImageWriter())

#save the barcode as an image file
barcode_filename=barcode_obj.save("ean13_barcode")

print(f"Barcode saved as{barcode_filename}.png")
```

```python
#this is used to open the image in your local pc. since this is colab it can't access the file explorer so use display
from PIL import Image
image=Image.open("ean13_barcode.png")
image.show()
```
