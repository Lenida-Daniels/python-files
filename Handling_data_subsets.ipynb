**Data Needed:** female people from Kenya whose age is 18 and above. Remember to do away with outliers

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data={
    "Age":[1,18,20,20,22,21,22,21,19,100],
    "country":["Kenya","Nigeria","Kenya","Uganda","Kenya","Kenya","Kenya","Tanzania","Kenya","Kenya"],
    "gender":["female","male","female","female","male","female","male","female","female","female"]
}
```

```python
df=pd.DataFrame(data)
df
```

```python
subset_gender=df["gender"]=="female" #only females
df[subset_gender]
   
```

```python
subset_country=df["country"]=="Kenya"#only Kenyans
df[subset_country]
```

```python
subset_age=df["Age"]>=18 #above 18
df[subset_age]
```

```python
Data_needed=df[subset_gender & subset_age & subset_country ]
Data_needed
```

visualization to identify outlier if present

```python
plt.hist(df["Age"]),
plt.xlabel("AGE",color="blue")
plt.title("AGE DISTRIBUTION",color="blue");
#100 is far away from the rest thus will lead to inaccuracy, so remove it.
```

```python
low,high=df["Age"].quantile([0.1,0.9]).round().astype(int)
remove_outlier=df["Age"].between(low,high)
df[remove_outlier]
```

```python
Data_needed=df[subset_gender & subset_age & subset_country &remove_outlier ]
Data_needed
```

```python
plt.hist(Data_needed["Age"]),
plt.xlabel("AGE",color="blue")
plt.xlim(18,23)
plt.title("AGE DISTRIBUTION",color="blue");
```
