```python
def patternA(rows):
    for i in range(rows):
        print("*"*i)
rows=5        
patternA(rows)
```

```python
def patternB(rows):
    for i in range(rows):
        for j in range(rows):
           print("*"*j)
    print()
rows=5        
patternB(rows)
```

```python
BDS=[45,25]
gender=['male','female']
col=['green','yellow']
exp=[0.1,0]
```

```python
import matplotlib.pyplot as plt
```

**PIE CHART**

```python
plt.figure(figsize=[4,4])
plt.pie(x=BDS,labels=gender,colors=col,explode=exp)
plt.title("Class Distribution")
plt.show();
```

```python
myd={"male":10,"female":20}
plt.pie(x=myd.values(),labels=myd.keys())
plt.title("AN EXAMPLE");
```

**BAR GRAPH**

```python
plt.bar(x=gender,height=BDS,color=col)
plt.title("DIstribution of the class");
```

```python
plt.barh(y=gender,width=BDS,color=col);
```

```python
day=[1,2,3,4,5,6,7]
loss=[15,26,30,45,36,50,34]
plt.plot(day,loss,marker="o",color="g")
plt.xlabel("day",color="b")
plt.ylabel(" Financial loss",color="b")
plt.title("FINANCIAL LOSS PER DAY",color="b");
```

**SCATTER PLOT**

```python
plt.scatter(day,loss)
plt.xlabel("day",color="b")
plt.ylabel(" Financial loss",color="b")
plt.title("FINANCIAL LOSS VS DAY",color="b");
```
