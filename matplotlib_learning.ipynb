```python
y1=np.array([1,2,3,4,5])
x1=np.array([0,4,6,8,12])
y2=np.array([1,4,7,9,10])
#linestyle can be written as ls
#dotted as :,dashed as --,solid(default) as -,dashdot as -.
#color can be written as c
#linewidth can be lw
plt.plot(y1,x1,'r-o',linestyle="-.",c='hotpink',lw=2)
plt.plot(y2)
plt.show()
```

```python
x=np.array([1,2,3,4,5])
y=np.array([5,10,15,20,25])
font1 = {'family':'serif','color':'blue','size':20}
plt.title("work per hour graph",loc="center",fontdict=font1)
plt.xlabel("hours")
plt.ylabel("work")
plt.plot(x,y)
plt.grid(color='green',linestyle='--',linewidth='0.5')
```
